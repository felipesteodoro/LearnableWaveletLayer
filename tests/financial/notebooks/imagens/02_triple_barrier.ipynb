# Figura 2: Método Triple Barrier

Visualização do Método Triple Barrier para geração de rótulos de classificação binária.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Configuração de estilos
plt.style.use('seaborn-v0_8-darkgrid')
fig, ax = plt.subplots(figsize=(12, 7))

# Simulação de caminho de preço com tendência
np.random.seed(42)
n_days = 50
t = np.arange(n_days)
trend = 0.002 * t
noise = np.random.normal(0, 0.015, n_days)
price = 100 * (1 + trend + np.cumsum(noise))

# Parâmetros TBM
volatility = np.std(np.diff(price) / price[:-1])
barrier_up = price[0] * (1 + 2.0 * volatility)
barrier_down = price[0] * (1 - 2.0 * volatility)
t_max = 10

# Identificar eventos
start_idx = 5
entry_price = price[start_idx]
time_barriers = [start_idx + t_max]

# Plotar série de preços
ax.plot(t, price, 'o-', color='#2c3e50', linewidth=2.5, markersize=4, label='Preço de fechamento')

# Linhas de barreira
ax.axhline(y=barrier_up, color='#e74c3c', linestyle='--', linewidth=2.5, label=f'Barreira superior ($k_u = 2.0$)')
ax.axhline(y=barrier_down, color='#27ae60', linestyle='--', linewidth=2.5, label=f'Barreira inferior ($k_d = 2.0$)')
ax.axvline(x=time_barriers[0], color='#f39c12', linestyle=':', linewidth=2.5, label=f'Barreira temporal ($T_{{max}} = 10$ dias)')

# Marcar ponto de entrada
ax.plot(start_idx, entry_price, '*', color='#2980b9', markersize=20, label='Ponto de entrada')

# Sombrear região de decisão
ax.axvspan(start_idx, time_barriers[0], alpha=0.1, color='#95a5a6')

# Anotações
ax.text(start_idx + 2, barrier_up + 0.5, 'COMPRA', fontsize=10, weight='bold', color='#e74c3c')
ax.text(start_idx + 2, barrier_down - 0.5, 'VENDA', fontsize=10, weight='bold', color='#27ae60')
ax.text(time_barriers[0] + 0.3, price[time_barriers[0] - 2], 'NEUTRO', fontsize=10, weight='bold', color='#f39c12')

ax.set_xlabel('Dias', fontsize=12, weight='bold')
ax.set_ylabel('Preço (R$)', fontsize=12, weight='bold')
ax.set_title('Método Triple Barrier para Geração de Rótulos de Classificação', fontsize=14, weight='bold', pad=20)
ax.legend(loc='upper left', fontsize=11, framealpha=0.95)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('/Users/fteodoro/Dropbox/Doutorado/Tese/figuras/fig_02_triple_barrier.pdf', dpi=300, bbox_inches='tight')
plt.savefig('/Users/fteodoro/Dropbox/Doutorado/Tese/figuras/fig_02_triple_barrier.png', dpi=300, bbox_inches='tight')
plt.show()

print('Figura salva: fig_02_triple_barrier.{pdf,png}')
```
