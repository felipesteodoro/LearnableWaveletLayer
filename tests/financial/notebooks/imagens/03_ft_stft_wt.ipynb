# Figura 3: Fourier, STFT e Wavelet Transform

Comparação de Transformada de Fourier, STFT e Wavelet Transform em série temporal.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import pywt
import warnings
warnings.filterwarnings('ignore')

# Gerar sinal composto (baixa + alta frequência)
fs = 100
t = np.linspace(0, 2, fs*2, endpoint=False)
f1, f2 = 5, 20
x = np.sin(2*np.pi*f1*t) + 0.5*np.sin(2*np.pi*f2*t)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# (1) Série temporal
axes[0, 0].plot(t, x, 'b-', linewidth=1.5)
axes[0, 0].set_title('Série Temporal (5 Hz + 20 Hz)', fontsize=12, weight='bold')
axes[0, 0].set_xlabel('Tempo (s)')
axes[0, 0].set_ylabel('Amplitude')
axes[0, 0].grid(True, alpha=0.3)

# (2) Transformada de Fourier
freqs = np.fft.fftfreq(len(x), 1/fs)
fft_x = np.fft.fft(x)
axes[0, 1].semilogy(freqs[:len(freqs)//2], np.abs(fft_x[:len(freqs)//2]), 'r-', linewidth=1.5)
axes[0, 1].set_title('Transformada de Fourier', fontsize=12, weight='bold')
axes[0, 1].set_xlabel('Frequência (Hz)')
axes[0, 1].set_ylabel('Magnitude')
axes[0, 1].grid(True, alpha=0.3)
axes[0, 1].set_xlim([0, 40])

# (3) STFT
f, t_stft, Sxx = signal.spectrogram(x, fs, nperseg=64)
pcm = axes[1, 0].pcolormesh(t_stft, f, 10*np.log10(Sxx + 1e-10), shading='auto', cmap='viridis')
axes[1, 0].set_title('Transformada STFT', fontsize=12, weight='bold')
axes[1, 0].set_xlabel('Tempo (s)')
axes[1, 0].set_ylabel('Frequência (Hz)')
axes[1, 0].set_ylim([0, 40])
plt.colorbar(pcm, ax=axes[1, 0], label='Potência (dB)')

# (4) Wavelet Transform (Scalogram)
scales = np.arange(1, 64)
coeffs, freqs_cwt = pywt.cwt(x, scales, 'morl', method='integrate_psi')
pcm2 = axes[1, 1].contourf(t, freqs_cwt, np.abs(coeffs), levels=20, cmap='viridis')
axes[1, 1].set_title('Transformada Wavelet Contínua', fontsize=12, weight='bold')
axes[1, 1].set_xlabel('Tempo (s)')
axes[1, 1].set_ylabel('Frequência (Hz)')
axes[1, 1].set_ylim([0, 40])
plt.colorbar(pcm2, ax=axes[1, 1], label='Coeficientes')

plt.tight_layout()
plt.savefig('/Users/fteodoro/Dropbox/Doutorado/Tese/figuras/fig_03_ft_stft_wt.pdf', dpi=300, bbox_inches='tight')
plt.savefig('/Users/fteodoro/Dropbox/Doutorado/Tese/figuras/fig_03_ft_stft_wt.png', dpi=300, bbox_inches='tight')
plt.show()

print('Figura salva: fig_03_ft_stft_wt.{pdf,png}')
```
