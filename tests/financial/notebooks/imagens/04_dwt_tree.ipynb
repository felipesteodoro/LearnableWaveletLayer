# Figura 4: Árvore de Decomposição DWT

Visualização da decomposição DWT em dois níveis com subbandas.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch
import warnings
warnings.filterwarnings('ignore')

fig, ax = plt.subplots(figsize=(14, 10))
ax.set_xlim(0, 10)
ax.set_ylim(0, 12)
ax.axis('off')

# Paleta de cores
color_input = '#cce5ff'
color_approx = '#c3ebf5'
color_detail = '#ffdea0'

def draw_box(ax, x, y, w, h, label, color, fontsize=10):
    box = FancyBboxPatch((x-w/2, y-h/2), w, h, boxstyle='round,pad=0.1',
                          edgecolor='black', facecolor=color, linewidth=2)
    ax.add_patch(box)
    ax.text(x, y, label, ha='center', va='center', fontsize=fontsize, weight='bold')

def draw_arrow(ax, x1, y1, x2, y2):
    arrow = FancyArrowPatch((x1, y1), (x2, y2), arrowstyle='->', 
                           mutation_scale=20, linewidth=2, color='black')
    ax.add_patch(arrow)

# Nível 0: Entrada
draw_box(ax, 5, 11, 2, 0.8, r'$\mathbf{X}^{(0)} \in \mathbb{R}^{40 \times 5}$', color_input, 11)
ax.text(5, 10.5, 'Entrada: sinal financeiro', ha='center', fontsize=9, style='italic')

# Nível 1: A1 e D1
draw_arrow(ax, 5, 10.6, 3.2, 9.4)
draw_arrow(ax, 5, 10.6, 6.8, 9.4)

draw_box(ax, 3.2, 9, 1.8, 0.7, r'$A^{(1)}$', color_approx, 11)
ax.text(3.2, 8.5, r'$\mathbb{R}^{20 \times 5}$', ha='center', fontsize=8)
ax.text(3.2, 8.1, 'Aprox. nível 1', ha='center', fontsize=8, style='italic')

draw_box(ax, 6.8, 9, 1.8, 0.7, r'$D^{(1)}$', color_detail, 11)
ax.text(6.8, 8.5, r'$\mathbb{R}^{20 \times 5}$', ha='center', fontsize=8)
ax.text(6.8, 8.1, 'Detalhe nível 1', ha='center', fontsize=8, style='italic')

# Nível 2: A2 e D2 (apenas de A1)
draw_arrow(ax, 3.2, 8.65, 1.5, 7.45)
draw_arrow(ax, 3.2, 8.65, 4.9, 7.45)

draw_box(ax, 1.5, 7, 1.6, 0.7, r'$A^{(2)}$', color_approx, 10)
ax.text(1.5, 6.4, r'$\mathbb{R}^{10 \times 5}$', ha='center', fontsize=8)
ax.text(1.5, 6.0, 'Aprox. nível 2', ha='center', fontsize=8, style='italic')

draw_box(ax, 4.9, 7, 1.6, 0.7, r'$D^{(2)}$', color_detail, 10)
ax.text(4.9, 6.4, r'$\mathbb{R}^{10 \times 5}$', ha='center', fontsize=8)
ax.text(4.9, 6.0, 'Detalhe nível 2', ha='center', fontsize=8, style='italic')

# Concatenação
ax.text(5, 5.3, 'Zero-Padding + Concatenação', ha='center', fontsize=11, weight='bold')
ax.text(5, 4.9, r'$[D^{(1)}, \text{pad}(D^{(2)}), \text{pad}(A^{(2)})] \rightarrow \mathbb{R}^{20 \times 15}$', 
        ha='center', fontsize=10)

# Setas para concatenação
draw_arrow(ax, 6.8, 8.65, 5.5, 5)
draw_arrow(ax, 4.9, 6.65, 5, 5)
draw_arrow(ax, 1.5, 6.65, 4.5, 5)

# Resultado final
draw_box(ax, 5, 3.5, 2.2, 0.8, r'$\mathbf{Z} \in \mathbb{R}^{20 \times 15}$', '#c8dff5', 11)
ax.text(5, 2.9, 'Entrada para Backbone Neural', ha='center', fontsize=9, style='italic')

draw_arrow(ax, 5, 4.5, 5, 3.9)

# Legenda de operações
ax.text(0.3, 1.8, 'Operações por nível:', fontsize=10, weight='bold')
ax.text(0.3, 1.3, r'Nível $i$: convolução depthwise + downsampling por fator 2', fontsize=9)
ax.text(0.3, 0.8, r'Filtros aprendíveis: $\mathbf{h}, \mathbf{g}$ (via QMF)', fontsize=9)
ax.text(0.3, 0.3, r'Daubechies-4 (fixo) ou LWT (aprendível)', fontsize=9)

# Título
fig.suptitle('Árvore de Decomposição Discreta de Wavelet em 2 Níveis', 
             fontsize=14, weight='bold', y=0.98)

plt.tight_layout()
plt.savefig('/Users/fteodoro/Dropbox/Doutorado/Tese/figuras/fig_04_dwt_tree.pdf', dpi=300, bbox_inches='tight')
plt.savefig('/Users/fteodoro/Dropbox/Doutorado/Tese/figuras/fig_04_dwt_tree.png', dpi=300, bbox_inches='tight')
plt.show()

print('Figura salva: fig_04_dwt_tree.{pdf,png}')
```
